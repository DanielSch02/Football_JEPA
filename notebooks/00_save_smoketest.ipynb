# 00 · Save smoke-test
Confirms `kaggle datasets create` works in YOUR Kaggle environment **before** the 13 GB run.
Creates a tiny throwaway dataset from a dummy folder. Run interactively (seconds).

**Settings:** Internet ON (accelerator None is fine).

## Make a tiny dummy folder + metadata

```python
import json, os
TEST = "/kaggle/working/save_test"
os.makedirs(TEST, exist_ok=True)
with open(f"{TEST}/hello.txt","w") as f:
    f.write("save mechanics test
")
json.dump({"title":"save-test-delete-me",
           "id":"daniels02/save-test-delete-me",
           "licenses":[{"name":"CC0-1.0"}]},
          open(f"{TEST}/dataset-metadata.json","w"))
print("dummy folder ready:", os.listdir(TEST))
```

## Try to create the dataset (this is the exact mechanic the big run uses)

```python
import subprocess
TEST = "/kaggle/working/save_test"
r = subprocess.run(["kaggle","datasets","create","-p",TEST],
                   capture_output=True, text=True)
print("STDOUT:", r.stdout)
print("STDERR:", r.stderr)
print("return code:", r.returncode)
```

## Interpret
- Success line / URL  -> ✅ save works; the big run's save will work too.
- `401/403` or 'credentials' -> Kaggle API not authed in this session (rare on Kaggle); tell Claude.
- 'already exists' -> it worked before; fine. You can delete `save-test-delete-me` from Your Datasets.
